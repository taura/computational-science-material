
#  応用編: 偏微分方程式 --- 差分法 (stencil)
# 1. この章のねらい
* 偏微分方程式を, 格子上の **近傍平均 (stencil)** の反復で解く差分法を扱う。
* 内側の二重ループを `collapse(2)` で並列化し, 収束判定の残差を `reduction(max)` で取る, という典型構造を学ぶ。
* GPU 化・台数効果の測定にもつながる。
