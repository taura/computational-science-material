
#  応用編: N体問題 (O(N²) の相互作用)
# 1. この章のねらい
* 多数の粒子が互いに力を及ぼし合う系 (重力, 群れ) を扱う。
* 各粒子について他の全粒子からの寄与を足し込む二重ループ (O(N²)) が主役で, 並列化・SIMD・GPU の効果が綺麗に見える。

# 2. 前提とする既習トピック
* `03_for_collapse` (多重ループ), `06_reduction` (力の集計)
* (発展) `11`–`13` (SIMD/ILP), `07`–`10` (GPU)

# 3. 含む問題 (準備中)
* `problems/00_nbody` --- 重力 N体 (直接法 O(N²))
* `problems/01_boids` --- Boids (群れの創発)

# 4. 到達目標
* O(N²) の相互作用計算を parallel for で並列化し, SIMD/GPU で加速できる。
