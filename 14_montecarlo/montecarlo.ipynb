
#  応用編: モンテカルロ法 (確率・期待値のシミュレーション)
# 1. この章のねらい
* ここからは「これまで学んだ技法 (parallel / for / reduction / GPU / SIMD) を使って, **意味のある計算**をする」応用編。
* この章では, 乱数を多数振って確率や期待値を推定する **モンテカルロ法** を扱う。
* 「たくさんの独立な試行を分担し, 結果を縮約 (reduction) でまとめる」という, 並列計算と相性の良いパターンが主役。

# 2. 前提とする既習トピック
* `02_parallel` (複数スレッドでの実行)
* `06_reduction` (総和・個数の縮約)。とくに `06_reduction/problems/02_montecarlo_pi` (モンテカルロでπ) が出発点。

# 3. 含む問題
* `problems/00_gacha` --- ガチャ (クーポンコレクター): 全種類そろえるまでの引き回数の期待値・ばらつき
* `problems/01_percolation` --- パーコレーション (浸透): ランダムな格子が「つながる」確率と相転移

# 4. 到達目標
* 独立試行を `parallel for` で分担し, `reduction` で平均・分散・個数を集計できる。
* シミュレーションの結果 (期待値, 確率, 相転移のしきい値) を「味わえる」形で出せる。
