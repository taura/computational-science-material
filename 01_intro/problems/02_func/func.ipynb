
# 練習問題: 関数定義と呼び出し

## 目標

関数の中身 (本体) を自分で書き, 引数を受け取って値を返す関数の仕組みを理解する.

## 課題

`func.cpp` (または `func.f90`) には, `x` を `n` 乗して返す関数 `power(x, n)` がある.
本体が空 (`p = 1` のまま) なので, 現状ではどんな入力でも `1` を返してしまう.

`TODO` の箇所に **`x` を `n` 回掛けるループ** を書き, `power` が正しく `x^n` を返すようにせよ.

- C++: `for (int i = 0; i < n; i++) { p = p * x; }`
- Fortran: `do i = 1, n` … `p = p * x` … `end do`

`main` 側 (関数の呼び出し) は変更しなくてよい.

## コンパイルと実行

```
# C++
nvc++ -fast func.cpp -o func.exe

# Fortran
nvfortran -fast func.f90 -o func.exe
```

```
./func.exe
```

## 期待される結果

```
2^10 = 1024.000000
3^4  = 81.000000
```

まず本体を空のまま実行すると両方 `1` になる (関数が呼べていることの確認). ループを書くと正しい値になる.
